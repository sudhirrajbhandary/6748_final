```python
import pandas as pd

def categorize_movie(row):
    profit = row['profit']
    roi = row['roi']
    if profit > 100000000 or roi > 4:
        return 'Blockbuster'
    elif roi > 0:
        return 'Success'
    else:
        return 'Flop'

# Load the dataset
#df = pd.read_csv("Imdb_Movie_Dataset.csv")

# Convert 'release_date' to datetime format
movies_revenue['release_date'] = pd.to_datetime(movies_revenue['release_date'], errors='coerce')

# Filter to include only movies released on or before June 1, 2023
cut_off_date = pd.to_datetime('2023-06-01')
movies_revenue = movies_revenue[(movies_revenue['release_date'] <= cut_off_date) & (movies_revenue['revenue'] > 0) & (movies_revenue['budget'] > 0)]

# Calculate Profit and ROI
movies_revenue['profit'] = movies_revenue['revenue'] - movies_revenue['budget']
movies_revenue['roi'] = (movies_revenue['profit'] / movies_revenue['budget']) #* 100  # ROI as a raw ratio

# Apply the categorize_movie function to the DataFrame
movies_revenue['success_level'] = movies_revenue.apply(categorize_movie, axis=1)

# Binning for Budget and Popularity
movies_revenue['budget_levels'] = pd.qcut(movies_revenue['budget'], 4, labels=["Low", "Medium", "High", "Very High"])
movies_revenue['popularity_tiers'] = pd.qcut(movies_revenue['popularity'], 4, labels=["Low", "Medium", "High", "Very Popular"])

# Count the number of movies after filtering
number_of_movies = movies_revenue.shape[0]
print("Number of movies after filtering:", number_of_movies)

# Save the enhanced dataset for further analysis
movies_revenue.to_csv("enhanced_movie_dataset.csv", index=False)

# Optionally, print the head of the DataFrame to check the new features
print(movies_revenue.head())
```
